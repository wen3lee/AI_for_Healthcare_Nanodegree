````python
dicom_019 = pydicom.dcmread('dicom_00023075_033.dcm')
dicom_019_image = dicom_019.pixel_array 
plt.imshow(dicom_019_image, cmap = 'gray')
#plt.imshow(dicom_019_image)
````

````python
plt.figure(figsize=(5,5))
plt.hist(dicom_019_image.ravel(), bins = 256)
plt.show()
````

````python
dicom_019_image
````

````python
dicom_019_image_mean = dicom_019_image.mean()
dicom_019_image_mean
````

````python
dicom_019_image_std = dicom_019_image.std()
dicom_019_image_std
````

````python
dicom_019_image_normalized = (dicom_019_image - dicom_019_image_mean)/dicom_019_image_std
dicom_019_image_normalized
````

````python
plt.figure(figsize=(5,5))
plt.hist(dicom_019_image_normalized.ravel(), bins = 256)
plt.show()
````
